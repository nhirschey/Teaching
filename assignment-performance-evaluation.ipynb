Isolate some notable portfolios.



````
type SignalPortfolioObs = 
    { Month: DateTime
      Name: string
      ExcessReturn: float }
let long =
    myExcessReturnPortfolios.Rows
    |> Seq.toList
    |> List.filter (fun row -> row.Index = Some 3)
    |> List.map (fun row -> 
        { Month = row.YearMonth
          Name = "Long"
          ExcessReturn = row.Ret })

let short =
    myExcessReturnPortfolios.Rows
    |> Seq.toList
    |> List.filter (fun row -> row.Index = Some 1)
    |> List.map (fun row -> 
        { Month = row.YearMonth
          Name = "Short"
          ExcessReturn = row.Ret })

let shortByMonth = 
    short 
    |> List.map (fun x -> x.Month, x) 
    |> Map

let longShort =
    [ for l in long do 
        // I'm indexing into shortByMonth directly
        // rather than using Map.tryFind because
        // I am pretty confident that the short 
        // data will be there
        let s = shortByMonth[l.Month]
        { Month = l.Month
          Name = "Long-short"
          ExcessReturn = l.ExcessReturn - s.ExcessReturn }]
````
